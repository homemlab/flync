# Predicting lncRNA

```python
import pickle
import pandas as pd
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier
script_path=Path.cwd()
appdir=script_path.parent.absolute()
outpath='/users3/chlab/rsantos/dev/fastq/dlinct'
model_path=appdir/'model'/'lncrna_model.sav'
prefix='new-lincRNAs'
predict_table=appdir/'test_get_features/features/new-lincRNAs/new-lincRNAs.csv'
```

```python
pt=pd.read_csv(predict_table)
pt.head()
```

```python
pt.drop('cov_me3', axis=1, inplace=True)
pt.drop('mean_pPcons27', axis=1, inplace=True)
```

```python
#Check duplicated values
pt.duplicated().value_counts()
```

```python
model_pt=pt.loc[:, 'length':'mean_pPcons124']
model_pt.head()
```

```python
## NEED TO FIX CONFLINCTING SCIKIT-LEARN VERSIONS
model=pickle.load(open(model_path, 'rb'))
```

```python
result=pd.DataFrame(model.predict(model_pt))
```

```python
result.rename(columns={0:'lncRNA'}, inplace=True)
result
```

```python
pos_result=result[result['lncRNA'] == True]
```

```python
pt_final=pd.concat([pt['name'], result], axis=1, verify_integrity=True)
```

```python
pt_pos=pd.concat([pt['name'], pos_result], axis=1, verify_integrity=True).dropna()
pt_pos
```

```python
pt_pos['name'].to_csv(outpath +'/'+prefix+'-names.csv', index=False, header=False)
pt_final.to_csv(outpath +'/'+prefix+'.csv', index=False, header=False)
```
